```python
# Set the x attributes and y attribute 

y=data['satisfaction']
X=data.drop('satisfaction',axis=1)
```

```python
X.columns.values
```

```python
#Features Scalling using standardisation
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
print(X)
```

```python
# Split the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

# Import library for modelling

```python
from sklearn.metrics import accuracy_score,classification_report,auc,roc_curve,confusion_matrix
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from math import sqrt
```

# Logistic Regression model

```python
# Train the logistic regression algorithm

from sklearn.linear_model import LogisticRegression 

LogR_model = LogisticRegression(random_state=1234) 
LogR_model.fit(X_train,y_train)
```

```python
# Apply for Test Set

y_pred_logR=LogR_model.predict(X_test)
y_pred_logR
```

```python
# Model Accuracy
from sklearn.metrics import accuracy_score
accLogR = accuracy_score(y_test,y_pred_logR)
accLogR_percent=accLogR*100
accLogR_percent=round(accLogR_percent,2)
print('Overall Accuracy score: ', accLogR_percent,'%')
```

```python
# Precision & Sensitivity/Recall Report

from sklearn.metrics import classification_report
print(classification_report(y_true=y_test,y_pred=y_pred_logR))
```

```python
import seaborn as sns

conf_matrix_logR = confusion_matrix(y_test, y_pred_logR)
print("Confusion Matrix for Logistic Regression:\n", conf_matrix_logR)
sns.heatmap(conf_matrix_logR, annot=None)
```

```python
# Compute ROC curve and AUC
fpr_logR, tpr_logR, thresholds_logR = roc_curve(y_test, y_pred_logR)
roc_auc_logR = auc(fpr_logR, tpr_logR)

# Print AUC score
print(f"AUC Score for Logistic Regression: {roc_auc_logR}")
```

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_logR, tpr_logR, color='orange', lw=2, label=f'AUC = {roc_auc_logR:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Logistic Regression')
plt.legend(loc="lower right")
plt.show()
```

```python
# Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
mse_logR = mean_squared_error(y_test, y_pred_logR)
rmse_logR = sqrt(mse_logR)

print(f"Mean Squared Error (MSE) for Logistic Regression: {mse_logR}")
print(f"Root Mean Squared Error (RMSE) for Logistic Regression: {rmse_logR}")
```

# K-Nearest Neighbors Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier

#Method to find the best value for k
error = []
for i in range(1,40):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  pred_i = knn.predict(X_test)
  error.append(np.mean(pred_i != y_test))

plt.figure(figsize=(12,6))
plt.plot(range(1, 40), error,
         color='red', linestyle='dashed',marker='o',
         markerfacecolor='blue',markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
best_k = np.argmin(error) + 1
print(f"Best k value: {best_k}")
```

Based solely on the error rate information presented in the graph, k = 0.5 is the optimal choice for minimizing the error for KNN model.

```python
from sklearn.neighbors import KNeighborsClassifier
#setting hyperparameter (optimal)
knn = KNeighborsClassifier(n_neighbors=7, metric='manhattan')
#train the model
knn.fit(X_train, y_train)
#apply to the test set
pred_knn = knn.predict(X_test)
#check the model accuracy
acc_knn = accuracy_score(y_test,pred_knn)
acc_knn
```

```python
report_knn = classification_report(y_test, pred_knn)

print("Classification Report for KNN:\n", report_knn)
```

```python
# Confusion Matrix
conf_matrix_knn = confusion_matrix(y_test, pred_knn)
print("Confusion Matrix for KNN:\n", conf_matrix_knn)
sns.heatmap(conf_matrix_knn, annot=None)
```

```python
# Compute ROC curve and AUC
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, pred_knn)
roc_auc_knn = auc(fpr_knn, tpr_knn)

# Print AUC score
print(f"AUC Score for KNN: {roc_auc_knn}")
```

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_knn, tpr_knn, color='orange', lw=2, label=f'AUC = {roc_auc_knn:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for KNN')
plt.legend(loc="lower right")
plt.show()
```

```python
# Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
mse_knn = mean_squared_error(y_test, pred_knn)
rmse_knn = sqrt(mse_knn)

print(f"Mean Squared Error (MSE) for KNN: {mse_knn}")
print(f"Root Mean Squared Error (RMSE) for KNN: {rmse_knn}")
```
